# üåÄ Python `for` Loop - A Beginner's Guide

## üìò What is a `for` loop?
A `for` loop in Python is used to iterate over a sequence like a list, tuple, dictionary, set, or string.

## üîÅ Syntax

```
for variable in sequence:
    # code block to execute
```

## üß™ Examples

### 1. Looping through a list

```
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

### 2. Looping through a string

```
for char in "hello":
    print(char)
```

### 3. Using `range()`

```
for i in range(5):
    print(i)
```

### 4. Using `else` with `for` loop

```
for i in range(3):
    print(i)
else:
    print("Loop finished!")
```

## ‚õî Using `break` to exit the loop

```
for i in range(10):
    if i == 5:
        break
    print(i)
```

## ‚Ü™Ô∏è Using `continue` to skip an iteration

```
for i in range(6):
    if i == 3:
        continue
    print(i)
```

## üìö Looping through a dictionary

```
student_scores = {"Alice": 85, "Bob": 92, "Charlie": 78}
for student, score in student_scores.items():
    print(f"{student} scored {score}")
```

## üí° Tips

```
colors = ["red", "green", "blue"]
for index, color in enumerate(colors):
    print(index, color)
```

```
names = ["Tom", "Jerry", "Spike"]
scores = [80, 90, 70]
for name, score in zip(names, scores):
    print(f"{name}: {score}")
```

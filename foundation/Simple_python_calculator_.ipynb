```python
# âœ… Install Gradio (if not already)
!pip install -q gradio

# âœ… Import Gradio
import gradio as gr

# âœ… Define the Calculator Function
def calculator(num1, num2, operation):
    try:
        num1 = float(num1)
        num2 = float(num2)
    except:
        return "Invalid input. Please enter numeric values."

    if operation == "Add":
        return num1 + num2
    elif operation == "Subtract":
        return num1 - num2
    elif operation == "Multiply":
        return num1 * num2
    elif operation == "Divide":
        if num2 == 0:
            return "Error: Division by zero"
        return num1 / num2
    else:
        return "Invalid Operation"

# âœ… Gradio Interface
calc_interface = gr.Interface(
    fn=calculator,
    inputs=[
        gr.Textbox(label="Enter first number"),
        gr.Textbox(label="Enter second number"),
        gr.Radio(["Add", "Subtract", "Multiply", "Divide"], label="Operation")
    ],
    outputs=gr.Textbox(label="Result"),
    title="ðŸ§® Simple Python Calculator",
    description="Perform basic arithmetic operations with this Gradio-powered calculator."
)

# âœ… Launch the App
calc_interface.launch(share=True)
```

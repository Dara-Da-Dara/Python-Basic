# Pandas Basics in Python

This notebook demonstrates basic Pandas operations with dummy data.

```python
###Pandas is a Python library designed for data manipulation and analysis.
###It provides powerful data structures such as Series and DataFrame to work with structured data effectively.
```

```python
import pandas as pd

# Create a dummy DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 55000, 70000]
}

df = pd.DataFrame(data)

# Display the DataFrame
print("DataFrame:")
print(df)
```

## Viewing Data
Use `head`, `columns`, and `shape` to inspect the DataFrame.

```python
# View the first rows
print(df.head())

# Display column names
print(df.columns)

# Shape of the DataFrame
print(df.shape)
```

## Adding and Modifying Data
Add new columns and manipulate existing data.

```python
# Add a new column for bonuses
df['Bonus'] = df['Salary'] * 0.1
print("\nDataFrame with Bonus:")
print(df)
```

## Filtering Data
Filter rows based on conditions.

```python
# Filter employees with salary greater than 55000
high_salary = df[df['Salary'] > 55000]
print("\nEmployees with Salary > 55000:")
print(high_salary)
```

## Grouping Data
Aggregate data by groups.

```python
# Group by Age and calculate average salary
print("\nAverage Salary by Age:")
print(df.groupby('Age')['Salary'].mean())
```

## Saving and Loading Data
Save DataFrame to CSV and reload it.

```python
# Save to CSV
df.to_csv("dummy_data.csv", index=False)

# Read the CSV
new_df = pd.read_csv("dummy_data.csv")
print("\nLoaded DataFrame:")
print(new_df)
```
